```python
def manh(p1,p2):
    distance=0
    for x1,x2 in zip(p1,p2):
        diff=x2-x1
        abs_diff=abs(diff)
        distance+=abs_diff
    return distance
```

```python
x1=(1,2,3,4,5,6)
x2=(10,20,30,1,2,3)
print(manh(x1,x2))
```
